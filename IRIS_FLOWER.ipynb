##TRAIN TEST SPLIT

````python
from sklearn.model_selection import train_test_split
train_test_split?
````

````python
df_train,df_tst=train_test_split(df,test_size=0.28)
print(df_train.shape,df_tst.shape)
df_train.head()
````

splitting data

````python
x_train=df_train.drop(columns=["target","target_names"]).values
y_train=df_train["target"].values
x_test=df_tst.drop(columns=["target","target_names"]).values
y_test=df_tst["target"].values
print(x_train.shape)
print(x_test.shape)
````

33% accuracy
95% accuracy explaination

````python
#x_train[1:4,2:4]
````

logistic regression

````python
from sklearn.linear_model import LogisticRegression
LogisticRegression?
model=LogisticRegression(max_iter=200) #it gives a convergence warning
#model.fit(x_train,y_train)
#model.score()
````

keeping test set for final test
###we split the traing set again for validation

````python
xt,xva,yt,yva=train_test_split(x_train,y_train,test_size=0.3)
model.fit(xt,yt)
````

````python
y_pr=model.predict(xva)
print(f"accuracy is {np.mean(y_pr==yva)*100 :.3f} %")
model.score(xva,yva)
````

USING CROSS VALIDATION
###explaination

````python
from sklearn.model_selection import cross_val_score, cross_val_predict
accuracies=cross_val_score(model,x_train,y_train,cv=5)
print(accuracies)
np.mean(accuracies)
````

modelling

````python
for k in {0.1,0.5,1,2,3,5,10,50,100}:
  model=LogisticRegression(max_iter=400,C=k)
  accuracies=cross_val_score(model,x_train,y_train,cv=5)
  #print(accuracies)
  print(str(k) +":"+ str(np.mean(accuracies)))
````
